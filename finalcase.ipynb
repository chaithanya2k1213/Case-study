Q3 Extract the rating as your target variable ‘y’ and all numerical parameters as your predictors ‘x’.
Separate 25% of your data as test set.

```python
df=cereal[['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups','rating']]
```

```python
df
```

```python
X=df.drop(['rating'],axis=1,inplace=False)
y=df['rating']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=4)
```

Q 4 Fit a linear regression module and measure the mean squared error on test dataset.

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
linear_model.fit(x_train,y_train) 
```

```python
#predicting the model
xpredicted_rating = linear_model.predict(x_test)
```

```python
from sklearn import metrics as mtr
print('Mean Square Error=',mtr.mean_squared_error(y_test,xpredicted_rating))
```

```python
plt.scatter(np.array(xpredicted_rating), np.array(y_test))
```

```python
from sklearn.metrics import r2_score
print( "r2 Score: ")
print (r2_score(y_test,xpredicted_rating))
```

```python
cereal_data=pd.read_csv('cereal.csv')
```

```python
cereal_data=cereal_data[['type','calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','weight','cups']]
```

```python
cereal_data
```

```python
x1=cereal_data.drop(['type'],axis=1)
```

```python
x1
```

```python
y1=cereal_data.type.values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.25,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(n_estimators=20,random_state=2)
```

```python
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
x_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test ,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
